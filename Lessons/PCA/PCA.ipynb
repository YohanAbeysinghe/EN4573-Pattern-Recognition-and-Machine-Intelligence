
# The Iris Dataset
This data sets consists of 3 different types of irises'
(Setosa, Versicolour, and Virginica) petal and sepal
length, stored in a 150x4 numpy.ndarray

The rows being the samples and the columns being:
Sepal Length, Sepal Width, Petal Length and Petal Width.

The below plot uses the first two features.
See [here](https://en.wikipedia.org/wiki/Iris_flower_data_set) for more
information on this dataset.

## Loading the iris dataset



```python
from sklearn import datasets

iris = datasets.load_iris()
```

## Scatter Plot of the Iris dataset



```python
import matplotlib.pyplot as plt

_, ax = plt.subplots()

# Define custom colors for each class
colors = ['red', 'green', 'blue']

# Create a list of colors based on the target labels
class_colors = [colors[label] for label in iris.target]

scatter = ax.scatter(iris.data[:, 0], iris.data[:, 1], c=class_colors)
ax.set(xlabel=iris.feature_names[0], ylabel=iris.feature_names[1])

# Create a custom legend using Line2D for each class and color
custom_legend_labels = [plt.Line2D([0], [0], marker='o', color=color, markersize=8, label=label) for color, label in zip(colors, iris.target_names)]
_ = ax.legend(handles=custom_legend_labels, loc="lower right", title="Classes")

plt.show()
```

## Plot a PCA representation
Let's apply a Principal Component Analysis (PCA) to the iris dataset
and then plot the irises across the first three PCA dimensions.
This will allow us to better differentiate between the three types!



```python
import mpl_toolkits.mplot3d  # noqa: F401
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import numpy as np

# Load the Iris dataset (or you can use your own dataset)
iris = load_iris()

fig = plt.figure(1, figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d", elev=-150, azim=110)

X_reduced = PCA(n_components=3).fit_transform(iris.data)

# Define custom colors for each class
colors = ['red', 'green', 'blue']

# Use custom colors based on the target labels
custom_colors = [colors[int(label)] for label in iris.target]



# Create a legend
legend_labels = [f'{label}: {color}' for label, color in zip(iris.target_names, colors)]
ax.legend(handles=[plt.Line2D([0], [0], marker='o', color='w', markersize=8, markerfacecolor=color, label=label) for label, color in zip(iris.target_names, colors)],
          
           loc="upper center",
           bbox_to_anchor=(0.75, 0.3))



ax.scatter(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2], c=custom_colors, s=40)

ax.set_title("First three PCA dimensions")
ax.set_xlabel("1st Eigenvector")
ax.xaxis.set_ticklabels([])
ax.set_ylabel("2nd Eigenvector")
ax.yaxis.set_ticklabels([])
ax.set_zlabel("3rd Eigenvector")
ax.zaxis.set_ticklabels([])



plt.show()
```

PCA will create 3 new features that are a linear combination of the
4 original features. In addition, this transform maximizes the variance.
With this transformation, we see that we can identify each species using
only the first feature (i.e. first eigenvalues).



Plot first and second PCA components




```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris

# Load the Iris dataset (or you can use your own dataset)
iris = load_iris()

# Perform PCA with 2 components (to project data into 2D)
X_reduced = PCA(n_components=2).fit_transform(iris.data)

# Define custom colors for each class
colors = ['red', 'green', 'blue']

# Create a scatter plot with custom colors for each class
plt.figure(figsize=(8, 6))
for i, color in enumerate(colors):
    class_indices = iris.target == i
    plt.scatter(X_reduced[class_indices, 0], X_reduced[class_indices, 1], c=color, label=iris.target_names[i], s=40)

plt.title("2D PCA Projection of Data")
plt.xlabel("1st Principal Component")
plt.ylabel("2nd Principal Component")
plt.legend()
plt.show()

```
