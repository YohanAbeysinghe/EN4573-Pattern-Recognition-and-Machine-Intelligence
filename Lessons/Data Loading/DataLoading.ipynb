```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme() #Applying the default theme
penguins = sns.load_dataset("penguins")
feature_names = penguins.columns[:-1] # Get the feature names (column names) from the DataFrame
class_names = penguins["species"].unique() # Get the unique class names (species)
# Print the feature names and class names
print("Feature Names:", feature_names)
print("Class Names:", class_names)
```

```python
# Plot the class distribution
plt.figure(figsize=(8, 6))
sns.countplot(data=penguins, x= "species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.title("Class Distribution of Penguins Species")
plt.show()
```

```python
print(penguins.info()) # Display information about the dataset
```

```python
# Get a list of numerical features
numerical_features = penguins.select_dtypes(include=['float64']).columns

# Create individual box plots for each numerical feature
for feature in numerical_features:
    plt.figure(figsize=(6, 4))
    sns.boxplot(data=penguins, x=feature)
    plt.xlabel(feature)
    plt.ylabel('Values')
    plt.title(f'Box Plot of {feature} in Penguins Dataset')
    plt.show()
```

```python
# Summary statistics for the penguins DataFrame
summary_stats = penguins.describe()

# Print the summary statistics
print(summary_stats)
```

```python
# Get a list of numerical features
numerical_features = penguins.select_dtypes(include=['float64']).columns

# Create individual box plots for each numerical feature
for feature in numerical_features:
    plt.figure(figsize=(6, 4))
    sns.boxplot(data=penguins.drop(columns="species"), x=feature)
    sns.stripplot(data=penguins.drop(columns="species"), x=feature, jitter=True, color="black", alpha=0.3)
    plt.xlabel(feature)
    plt.ylabel('Values')
    plt.title(f'Box Plot of {feature} in Penguins Dataset')
    plt.show()
```

```python
# Calculate covariance matrix and variance for each variable
cov_matrix = penguins.cov()
variance_values = penguins.var()

# Create the pairplot with covariance plots
g = sns.pairplot(data=penguins, hue="species", diag_kind="kde")
plt.show()
```
